```python
#Dropping irrelevant columns
df.drop(columns={'Employee ID','PayZone','DOB'},inplace=True)
```

```python
#Renaming columns
df=df.rename(columns={"Title":"Job Title","EmployeeType":"Employment Type","EmployeeStatus":"Status","EmployeeClassificationType":"Work Status",
                      "DepartmentType": "Department","GenderCode":"Gender","RaceDesc":"Ethnicity",
                      "MaritalDesc":"Marital Status","Training Program Name":"Training Program","BusinessUnit":"Business Unit"})
df.head()
```

```python
#This step involves removing any trailing spaces from text columns.
cat_cols=df.select_dtypes(include=object)
for col in cat_cols:
    df[col]=df[col].str.strip()
```

```python
# Converting date columns to datetime format
date_cols=['StartDate','Survey Date','Training Date']
for col in date_cols:
    df[col]=pd.to_datetime(df[col],format='mixed')
df[date_cols].head()    
```

```python
df[date_cols].info()
```

```python
#Inspecting Unique Values in Categorical Columns
cat_cols=df.select_dtypes(include=object)
for col in cat_cols:
    print(f"\nColumn: {col}")
    print(f"Unique values ({df[col].nunique()}): {df[col].unique()}")
    print("-" * 40)
```

* To clean the data, we will group similar values under a single job title.

```python
df['Job Title']=df['Job Title'].str.replace(r'.*Production Technician.*', 'Production Technician', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Network Engineer*', 'Network Engineer', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*IT Manager.*', 'IT Manager', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Sr. DBA*', 'DBA', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Database Administrator*', 'DBA', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Data Architect.*', 'Data Architect', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*BI Developer.*', 'BI Developer', regex=True)
df['Job Title'] = df['Job Title'].str.replace(r'.*IT Manager.*', 'IT Manager', regex=True)
```

* We are doing the same thing with the divisions.

```python
df['Division']=df['Division'].str.replace(r'.*General.*', 'General', regex=True)
df['Division']=df['Division'].str.replace(r'.*Project Management.*', 'Project Management', regex=True)
```

* For better data understanding, we will replace state abbreviations with full names.

```python
state = {
    'MA': 'Massachusetts', 'ND': 'North Dakota', 'FL': 'Florida', 'CT': 'Connecticut', 'CA': 'California',
    'OR': 'Oregon', 'TX': 'Texas', 'IN': 'Indiana', 'GA': 'Georgia', 'CO': 'Colorado', 'KY': 'Kentucky',
    'NV': 'Nevada', 'VT': 'Vermont', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'VA': 'Virginia',
    'UT': 'Utah', 'ID': 'Idaho', 'MT': 'Montana', 'WA': 'Washington', 'AZ': 'Arizona', 'NH': 'New Hampshire',
    'AL': 'Alabama', 'OH': 'Ohio', 'NY': 'New York', 'ME': 'Maine', 'TN': 'Tennessee', 'NC': 'North Carolina'
}
df['State'] = df['State'].map(state)
```

```python
cat_cols=df.select_dtypes(include=object)
print('Count of unique values')
print("-" * 40)
for col in cat_cols:
    print(f"{df[col].value_counts()}")
    print("-" * 40)
```

* Correlation between numerical features

```python
corr=df.select_dtypes(include='number')
plt.figure(figsize=(12,6))
sns.heatmap(data=corr.corr(),annot=True)
plt.show()
```

* To improve visualization, we'll extract year and month from the 'Start Date' column.

```python
df['Year']=df['StartDate'].dt.year
df['Month']=df['StartDate'].dt.month
df.head()
```

### 3. Data Visualization

```python
year_df=df.groupby('Year')['Year'].size().reset_index(name='count')
label_max=year_df['count'].max()
label_min=year_df['count'].min()
plt.figure(figsize=(8,4))
sns.lineplot(data=year_df,x='Year',y='count',color='blue')
plt.annotate(label_max,xy=(2022,600),color='green')
plt.annotate(label_min,xy=(2018,240),color='red')
plt.annotate('', xy=(2019, 560),xytext=(2018, 255), xycoords='data', arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='red', lw=3))
plt.title('Number of Employments Through Years',fontsize=12)
plt.xlabel('Year',fontsize=10)
plt.ylabel('Total Employees',fontsize=10)
plt.show()
```

* From this line plot, we can see that the company hired the most employees in **2022**, with a total of **594**.
In 2018, the total number of hired employees was **248**, but there is a **sudden increase** from **2018 to 2019**, which could indicate that the company started in 2018.

```python
month_df=df.groupby('Month')['Month'].size().reset_index(name='count')
plt.figure(figsize=(8,4))
sns.lineplot(data=month_df,x='Month',y='count')
plt.title('Number of Employments Through Months',fontsize=12)
plt.xlabel('Month',fontsize=10)
plt.ylabel('Total Employees',fontsize=10)
plt.xticks(ticks=range(1,13))
plt.show()
```

* This line plot shows that company mostly hires in winter, between **November and December**, and at the start of spring, in **March**.

```python
attrition=df[df['Status']=='Terminated'].groupby('Year')['Status'].size().reset_index(name='count')
plt.figure(figsize=(8,4))
sns.lineplot(data=attrition,x='Year',y='count', label='Terminated Employees',marker='o')
plt.title('Number of Terminated Emoloyees Through Years',fontsize=12)
plt.xlabel('Year',fontsize=10)
plt.ylabel('Total Employees',fontsize=10)
plt.show()
```

* We can see that number of terminated employees was the highest in **2019**, with the sudden decrease from **2022 to 2023**.

```python
status = df[df['Status']=='Terminated'].groupby(['Department', 'Status']).size().reset_index(name='Count')
plt.figure(figsize=(8, 4))
sns.lineplot(data=status,x='Department',y='Count', label='Terminated Employees')
plt.ylabel('Total Employees',fontsize=12)
plt.xlabel('')
plt.title('Number of Terminated Employees by Department',fontsize=14)
plt.show()
```

* The highest number of employee terminations occurred in the **Production Department**, which could indicate a poor work environment leading to high turnover, or suggest that the company was intentionally reducing staff in that area.

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Age'],kde=True)
plt.title('Age Distribution',fontsize=14)
plt.xlabel('Age',fontsize=12)
plt.ylabel('')
plt.show()
```

* Most employees are between **50 and 70** years old. This could be a wake-up call for the company to assess employee attrition and consider attracting younger talent.

```python
status = df[df['Status']=='Terminated'].groupby(['Age', 'Status']).size().reset_index(name='Count')
plt.figure(figsize=(8,4))
sns.lineplot(data=status,x='Age',y='Count', label='Terminated Employees')
plt.fill_between(status['Age'], status['Count'], alpha=0.3, color='red')
plt.title('Number of Terminated Employees by Age',fontsize=14)
plt.xlabel('Age',fontsize=12)
plt.ylabel('')
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10, 8))
gender_count=df.groupby('Gender')['Gender'].count().plot.pie(autopct='%1.1f%%',ax=axs[0,0],ylabel='')
axs[0, 0].set_title('Gender',fontsize=12)
marital_count=df.groupby('Marital Status')['Marital Status'].count().plot.pie(autopct='%1.1f%%',ax=axs[0,1],ylabel='')
axs[0, 1].set_title('Martial Status',fontsize=12)
ethnicity_count=df.groupby('Ethnicity')['Ethnicity'].count().plot.pie(autopct='%1.1f%%',ax=axs[1,0],ylabel='')
axs[1, 0].set_title('Ethnicity',fontsize=12)
sns.countplot(x='State', data=df[df['State'].isin(df['State'].value_counts().nlargest(5).index)],ax=axs[1,1],hue='State',legend=False)
plt.xticks(rotation=45)
plt.xlabel('State',fontsize=12)
plt.ylabel('')
fig.suptitle("Employee Demographics")

plt.tight_layout()
plt.show()
```

* The demographics show that most employees are based in **Massachusetts**, with a total of **2,500**. Gender distribution is quite balanced, with a slightly higher number of females.
Regarding ethnicity and marital status, the values are fairly evenly distributed.

```python
columns=['Department','Job Title','Division','Business Unit']
for col in columns:
    plt.figure(figsize=(8, 5))
    data=df.groupby(col)[col].size().reset_index(name='count')
    data=data.sort_values(by='count',ascending=False).head(10)
    sns.barplot(data=data,x='count',y=col,hue=col,legend=False,palette='tab10')
    plt.title(f'Distribution of {col}',fontsize=12)
    plt.xlabel('Total',fontsize=10)
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

* From these bar plots, we can see that the majority of employees work in the **Production Department** as **Production Technicians**.
Regarding divisions, most employees are in **Field Operations** and **General**. Employees are evenly distributed across all business units.

```python
columns=['Employment Type','Work Status','Performance Score']
for col in columns:
    plt.figure(figsize=(8, 5))
    data=df.groupby(col)[col].size().reset_index(name='count')
    data=data.sort_values(by='count',ascending=False).head(10)
    sns.barplot(data=data,x='count',y=col,hue=col,legend=False)
    plt.title(f'Distribution of {col}',fontsize=12)
    plt.xlabel('Total',fontsize=10)
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

* These bar plots show that employment type and work status are fairly evenly distributed. 
When it comes to performance scores, most employees fully meet expectations.

```python
data = df.groupby('Year')[['Engagement Score', 'Satisfaction Score', 'Work-Life Balance Score']].mean()
plt.figure(figsize=(10, 4))
sns.lineplot(data=data,markers=True)
plt.title('Average Employee Satisfaction Metrics by Year',fontsize=14)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Average Score',fontsize=12)
plt.show()
```

* All satisfaction metric scores are around **3.0**, which suggests there is room for improvement in the coming years.

```python
outcome = df.groupby('Training Program')['Training Outcome'].value_counts()
programs = outcome.index.get_level_values(0).unique()
legend_labels = outcome.index.get_level_values(1).unique()
fig, axs = plt.subplots(3, 2, figsize=(12, 8))
axs = axs.flatten()

for i, program in enumerate(programs):
    program_data = outcome[program]
    axs[i].pie(program_data, labels=None, autopct='%1.1f%%', startangle=90)
    axs[i].set_title(f'{program}',fontsize=12)
    axs[i].axis('equal')
for j in range(len(programs), len(axs)):
    axs[j].axis('off')

fig.legend(labels=legend_labels, loc='center right', title='Training Outcome',fontsize=12)
plt.suptitle('Training Outcome')
plt.tight_layout()
plt.show()
```

* The pie charts show that for each training program, about 50% of employees completed and passed, while the other 50% either didnt complete or failed the training.

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df,x='Training Program',y='Training Duration(Days)',hue='Training Program')
plt.ylabel('Days',fontsize=12)
plt.xlabel('',fontsize=12)
plt.title('Average Training Duration',fontsize=14)
plt.show()
```

* The average duration of each training program was around **3 days**, which might explain why many employees failed or didnt complete the training

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,x='Training Program',y='Training Cost',hue='Training Program')
plt.ylabel('Price',fontsize=12)
plt.xlabel('',fontsize=12)
plt.title('Average Price for Training Program',fontsize=14)
plt.show()
```

```python
df['Days to Train'] = (df['Training Date'] - df['StartDate']).dt.days

plt.figure(figsize=(8,4))
sns.histplot(df['Days to Train'], bins=30, kde=True)
plt.title('Days Between Start and Training',fontsize=14)
plt.xlabel('Days to Training',fontsize=12)
plt.ylabel('Number of Employees',fontsize=12)
plt.xticks(ticks=range(-300,1700,300))
plt.show()
```

**Negative values** in this plot might indicate that some training programs were completed before employees officially started working at the company. 
Many values fall between **1200 and 1500 days** (around 4 years), which could suggest that training was not initially required or that employees attended training later to gain additional skills.

### 4. Conclusion

* Looking at the data, its clear that the company hired the most employees in **2022**, with a big jump in hiring from **2018 to 2019**. This might indicate that the company started or began growing quickly around that time. Most hiring happens in the winter months, especially in **November and December**, and also in **March**.
* Employee terminations peaked in **2019**, especially in the **Production Department**. This could point to some challenges in that area, such as a tough work environment or a decision to reduce staff. Its also worth noting that most employees are between **50 and 70 years old**, which suggests the company might want to focus on bringing in younger talent for long-term sustainability.
* The majority of employees are based in **Massachusetts**, and the gender split is pretty balanced, with slightly more women. Ethnicity and marital status are also fairly evenly spread. Most people work in the **Production Department** as **Production Technicians**, and the biggest divisions are **Field Operations** and **General**. Employees are spread out fairly evenly across business units, and theres a good mix when it comes to work status and employment type.
* Most employees are meeting performance expectations, but satisfaction scores are only around **3**, showing theres room to improve how employees feel about their jobs.
* Training data shows that only about half of employees complete and pass the training programs. With each program lasting just around three days, that could explain the lower success rates. Also, some unusual negative values in the data suggest that a few people might have completed training before officially starting their jobs. Many employees seem to have done training much later on, possibly to learn new skills after they were already working.
* Overall, the company seems stable and growing, but there are clear areas to improveespecially in employee satisfaction, training effectiveness, and planning for future talent.
