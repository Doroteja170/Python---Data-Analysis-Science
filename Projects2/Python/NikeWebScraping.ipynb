```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Function to extract Product Title
def get_title(soup):

    try:
        title = soup.find(attrs={"id": "pdp_product_title"}).text.strip()
    # If there is no value, return empty string
    except AttributeError:
        title = ""

    return title

# Function to extract Product Price
def get_price(soup):

    try:
        price=soup.find("span",attrs={"class": "nds-text mr2-sm css-tbgmka e1yhcai00 text-align-start appearance-body1Strong color-primary display-inline weight-regular"}).text.strip()

    except AttributeError:
            price = ""

    return price


# Function to extract Product Type
def get_description(soup):

    try:
        description=soup.find(attrs={"id":"pdp_product_subtitle"}).text.strip()
    
    except AttributeError:
            description = ""	

    return description

# Function to extract Product Color
def get_color(soup):
    try:
        color = soup.find("li", {"data-testid": "product-description-color-description"}).get_text(strip=True).replace("Shown:", "").strip()

    except AttributeError:
        color = ""	

    return color
```

```python
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})

# Page URL
URL = "https://www.nike.com/w/womens-shoes-5e1x6zy7ok"

# HTTP Request
page = requests.get(URL, headers=HEADERS)
soup=BeautifulSoup(page.content,'html.parser')

# Extract all anchor elements into a list of Tag objects
links=soup.find_all("a",attrs={'class':'product-card__link-overlay'})

# Add the links
link_list = []

# Loop for extracting links from Tag Objects
for link in links:
    link_list.append(link.get('href')) 

# Data Dictionary
d = {"Product": [], "Price": [],"Type": [], "Color": []}

# Loop for extracting product details from each link 
for link in link_list: 
    new_page = requests.get(link, headers=HEADERS)  
    new_soup = BeautifulSoup(new_page.content, 'html.parser')  

# Function calls to display all product information
    d['Product'].append(get_title(new_soup))
    d['Price'].append(get_price(new_soup))
    d['Type'].append(get_description(new_soup))
    d['Color'].append(get_color(new_soup))

df = pd.DataFrame.from_dict(d)
df.to_csv("nikedata.csv",header=True)
```
